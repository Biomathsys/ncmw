# Hierarchical model compostion

This implements a way to compute Hierarchical metabolic models, according to the SBML "comp" package. This has several advantages:
* Within the SBML file only variables and constraints relevant for the community are saved. All other necessary variables stay within the original SBML files.
* Changes to the individual models will be automatically transfered to the community model.

The constructed SBML file does not contain the discription of the individual models, but instead only describes how these models should be combined! As a result we save there only things like shuttle reactions.

```python
from ncmw.community import HierarchicalCommunityModel
```

```python
model = HierarchicalCommunityModel(["../data/models/iDPM21RW.xml", "../data/models/iYS854.xml"])
```

Unfortunattly, hierarchical models are not supported within cobrapy. Yet, there exist an coverter the "flattening" algorithm. It does read the "description" on how to combine the individual models and exactly does this putting everythink into one file. This can now be read by cobrapy. All this is done in the background, in the end we obtain the same commuity model as before.

```python
model.optimize()
```

```python
model.cooperative_tradeoff(0.8)
```

```python
model.summary(cooperative_tradeoff=0.8)
```
